```python
import pandas as pd
import numpy as np

# Correct file path handling
reviews_datasets = pd.read_csv(r"C:\Users\quant\Downloads\Reviews.csv")

# Select the first 20,000 rows
reviews_datasets = reviews_datasets.head(20000)

# Drop NaN values and update the DataFrame
reviews_datasets = reviews_datasets.dropna()
```

```python
reviews_datasets['Text'][350]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(max_df=0.8, min_df=2, stop_words='english')
doc_term_matrix = count_vect.fit_transform(reviews_datasets['Text'].values.astype('U'))
```

```python
print(doc_term_matrix)




```

```python
from sklearn.decomposition import LatentDirichletAllocation

 

LDA = LatentDirichletAllocation(n_components=5, random_state=42) # 5 topics

LDA.fit(doc_term_matrix)




```

```python
# print 10 random words

import random

 

for i in range(10):

    random_id = random.randint(0,len(count_vect.get_feature_names_out()))

    print(count_vect.get_feature_names_out()[random_id])




```

```python
first_topic = LDA.components_[0] # topic 0




```

```python
top_topic_words = first_topic.argsort()[-10:] # last 10 words




```

```python
for i in top_topic_words:

    print(count_vect.get_feature_names_out()[i])




```

```python
for i,topic in enumerate(LDA.components_):

    print(f'Top 10 words for topic #{i+1}:')

    print([count_vect.get_feature_names_out()[i] for i in topic.argsort()[-10:]])

    print('\n')




```

```python
topic_values = LDA.transform(doc_term_matrix)

topic_values.shape




```

```python
# adding topic number to original data

reviews_datasets['Topic'] = topic_values.argmax(axis=1)




```

```python
import pandas as pd

import numpy as np

 

reviews_datasets = pd.read_csv(r"C:\Users\quant\Downloads\Reviews.csv")

reviews_datasets = reviews_datasets.head(20000)

reviews_datasets.dropna()




```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

 

tfidf_vect = TfidfVectorizer(max_df=0.8, min_df=2, stop_words='english')

doc_term_matrix = tfidf_vect.fit_transform(reviews_datasets['Text'].values.astype('U'))




```

```python
from sklearn.decomposition import NMF

 

nmf = NMF(n_components=5, random_state=42) # 5 topics

nmf.fit(doc_term_matrix)




```

```python
import random

 

for i in range(10):

    random_id = random.randint(0,len(tfidf_vect.get_feature_names_out()))

    print(tfidf_vect.get_feature_names_out()[random_id])




```

```python
# getting top 10 words from topic 0

first_topic = nmf.components_[0]

top_topic_words = first_topic.argsort()[-10:]




```

```python
for i in top_topic_words:

    print(tfidf_vect.get_feature_names_out()[i])




```
